```python
#Capital Value Rules

capital_value=[]
for row_pos in range(len(pnl_df)):
    row=pnl_df.iloc[row_pos,:]
    row_buy_date=row['BuyDate_new']
    row_sell_date=row['SellDate_new']

    row_buy_amt=row['BuyAmount']
    row_sell_amt=row['SellAmount']

    if row_sell_date>=row_buy_date:
        capital_value.append(row_sell_amt-row_buy_amt)
    else:
        capital_value.append(row_buy_amt-row_sell_amt)

pnl_df['Capital_Value']= capital_value

# Attribution_Month
pnl_df['PNL Attribution Month']=pnl_df.apply(lambda row: row['SellDate_new'].strftime('%Y-%m')
                                         if row['SellDate_new'] >= row['BuyDate_new'] else row['BuyDate_new'].strftime('%Y-%m'), axis=1)

#Pivoting and formatting as per required format
segments=['Equity','Derivatives','Commodity','Currency','MutualFund']
segment_dfs=[]
segment_pivot_dfs=[]
segment_pivot_trimmed_dfs=[]
segment_pivot_engineered_dfs=[]

month_last_num_list=[3,6,9,12]
month_last_list=["last 3 months","last 6 months","last 9 months","last 12 months"]
month_lists=[last_3_month_list,last_6_month_list,last_9_month_list,last_12_month_list]
month_lists_excl=[last_3_month_list_excl,last_6_month_list_excl,last_9_month_list_excl,last_12_month_list_excl]

month_last_dict=dict(zip(month_last_list,month_lists))
month_last_dict_excl=dict(zip(month_last_list,month_lists_excl))

for segment in segments:
    seg_df=pnl_df[pnl_df['Segment']==segment]

    segment_dfs.append(seg_df)
    seg_df_working=seg_df[['ClientCode','Capital_Value','PNL Attribution Month']]
    seg_df_working_grouped=seg_df_working.groupby(['ClientCode','PNL Attribution Month']).sum().reset_index()
    seg_df_working_grouped_pivot=seg_df_working_grouped.pivot(index='ClientCode',columns='PNL Attribution Month',values='Capital_Value').reset_index().fillna(0)

    suffix='_'+segment.lower()+'_pnl'
    seg_df_working_grouped_pivot=seg_df_working_grouped_pivot.add_suffix(suffix)
    seg_df_working_grouped_pivot.rename(columns={'ClientCode'+suffix:'ClientCode','PNL Attribution Month'+suffix:'PNL Attribution Month'},inplace=True)

    segment_pivot_dfs.append(seg_df_working_grouped_pivot)

    trimmed_col_list=['ClientCode']

    for col in seg_df_working_grouped_pivot.columns:
        col_yr_mth=col[0:7]
        if segment=='MutualFund':
            if col_yr_mth in month_list_excl:
                trimmed_col_list.append(col)
        else:
            if col_yr_mth in month_list:
                trimmed_col_list.append(col)

    seg_df_working_grouped_pivot_trimmed=seg_df_working_grouped_pivot[trimmed_col_list]
    segment_pivot_trimmed_dfs.append(seg_df_working_grouped_pivot_trimmed)

    # getting overall info
    pnl_cols=trimmed_col_list[1:]

    seg_df_working_grouped_pivot_trimmed[suffix+'_delta_months']= (seg_df_working_grouped_pivot_trimmed[pnl_cols] !=0).sum(axis=1)
    seg_df_working_grouped_pivot_trimmed[suffix+'_total']=seg_df_working_grouped_pivot_trimmed[pnl_cols].sum(axis=1)

    #seg_df_working_grouped_pivot_trimmed.dropna(inplace=True)
    #seg_df_working_grouped_pivot_trimmed.drop_duplicates(inplace=True)

    for month_last_pos in range(len(month_last_list)):
        month_last_num=month_last_num_list[month_last_pos]
        month_last=month_last_list[month_last_pos]

        month_list=month_last_dict[month_last]
        month_list_excl=month_last_dict_excl[month_last]

        engineered_col_list=['ClientCode']

        for col in seg_df_working_grouped_pivot_trimmed.columns:
            col_yr_mth=col[0:7]
            if segment=='MutualFund':
                if col_yr_mth in month_list_excl:
                    engineered_col_list.append(col)
            else:
                if col_yr_mth in month_list:
                    engineered_col_list.append(col)

        seg_df_working_grouped_pivot_engineered=seg_df_working_grouped_pivot_trimmed[engineered_col_list]
        pnl_cols=engineered_col_list[1:]

        #last 3,6,9,12m PnL
        seg_df_working_grouped_pivot_trimmed[month_last+suffix]= (seg_df_working_grouped_pivot_trimmed[pnl_cols]).sum(axis=1)

    seg_df_working_grouped_pivot_engineered=seg_df_working_grouped_pivot_trimmed
    segment_pivot_engineered_dfs.append(seg_df_working_grouped_pivot_engineered)




segment_df_dict=dict(zip(segments,segment_dfs))
segment_pivot_df_dict=dict(zip(segments,segment_pivot_dfs))
segment_pivot_df_trimmed_dict=dict(zip(segments,segment_pivot_trimmed_dfs))
segment_pivot_df_engineered_dict=dict(zip(segments,segment_pivot_engineered_dfs))

## Merging and Sorting Segments
pnl_engineered_segment_merged = reduce(lambda  left,right: pd.merge(left,right,on=['ClientCode'],
                                            how='outer'), segment_pivot_engineered_dfs)
pnl_engineered_segment_merged.head()

pnl_engineered_segment_merged_sorted = pnl_engineered_segment_merged.reindex(sorted(pnl_engineered_segment_merged.columns), axis=1)
cc_list=pnl_engineered_segment_merged_sorted['ClientCode']
pnl_engineered_segment_merged_sorted.pop('ClientCode')
pnl_engineered_segment_merged_sorted.insert(0, 'ClientCode', cc_list)

#Datework
month_list=[]
pay_in_out_month_list=[]

start=datetime(year=2018, month=1, day=1).date() #Start with Fy 22
end=datetime(year=2022, month=1, day=1).date()

while (start<end):
    month_list.append(start.strftime('%Y-%m'))
    start=start+relativedelta(months=1)

last_3_month_list=month_list[-3:]
last_6_month_list=month_list[-6:]
last_9_month_list=month_list[-9:]
last_12_month_list=month_list[-12:]

month_list

pnl_engineered_segment_merged = reduce(lambda  left,right: pd.merge(left,right,on=['ClientCode'],
                                            how='outer'), segment_pivot_trimmed_dfs)
pnl_engineered_segment_merged_sorted = pnl_engineered_segment_merged.reindex(sorted(pnl_engineered_segment_merged.columns), axis=1)
cc_list=pnl_engineered_segment_merged_sorted['ClientCode']
pnl_engineered_segment_merged_sorted.pop('ClientCode')
pnl_engineered_segment_merged_sorted.insert(0, 'ClientCode', cc_list)
```

```python
#Pivoting and formatting as per required format
segments=['Equity','Derivatives','Commodity','Currency']

segment_dfs=[]
segment_pivot_dfs=[]
segment_pivot_trimmed_dfs=[]

total_cols=[]

for segment in segments:
    seg_df=ledger_bal_bal_amt[ledger_bal_bal_amt['ExchType']==segment]
    segment_dfs.append(seg_df)
    seg_df_working=seg_df[['ClientCode','TotalBalamt','Quarter_Dormancy']]
    seg_df_working_grouped=seg_df_working.groupby(['ClientCode','Quarter_Dormancy']).sum().reset_index()
    seg_df_working_grouped_pivot=seg_df_working_grouped.pivot(index='ClientCode',columns='Quarter_Dormancy',values='TotalBalamt').reset_index().fillna(0)

    suffix='_'+segment.lower()+'_ledger_bal_total_bal'
    seg_df_working_grouped_pivot=seg_df_working_grouped_pivot.add_suffix(suffix)
    seg_df_working_grouped_pivot.rename(columns={'ClientCode'+suffix:'ClientCode','Quarter_Dormancy'+suffix:'Quarter_Dormancy'},inplace=True)
    segment_pivot_dfs.append(seg_df_working_grouped_pivot)

    trimmed_col_list=['ClientCode']
    for col in seg_df_working_grouped_pivot.columns:
        col_yr=col[0:4]
        if col_yr in ["2018","2019","2020","2021"]:
            trimmed_col_list.append(col)
    seg_df_working_grouped_pivot_trimmed=seg_df_working_grouped_pivot[trimmed_col_list]
    segment_pivot_trimmed_dfs.append(seg_df_working_grouped_pivot_trimmed)


segment_df_dict=dict(zip(segments,segment_dfs))
segment_pivot_df_dict=dict(zip(segments,segment_pivot_dfs))
segment_pivot_df_trimmed_dict=dict(zip(segments,segment_pivot_trimmed_dfs))

ledger_bal_bal_amt_merged = reduce(lambda  left,right: pd.merge(left,right,on=['ClientCode'],
                                            how='outer'), segment_pivot_dfs)
ledger_bal_bal_amt_merged_sorted = ledger_bal_bal_amt_merged.reindex(sorted(ledger_bal_bal_amt_merged.columns), axis=1)
cc_list=ledger_bal_bal_amt_merged_sorted['ClientCode']
ledger_bal_bal_amt_merged_sorted.pop('ClientCode')
ledger_bal_bal_amt_merged_sorted.insert(0, 'ClientCode', cc_list)

ledger_bal_bal_amt_merged_sorted['_ledger_bal_total_bal_avg']= ledger_bal_bal_amt_merged_sorted[total_cols].mean(axis=1)


segment_dfs=[]
segment_pivot_dfs=[]
segment_pivot_trimmed_dfs=[]

total_cols=[]

for segment in segments:
    seg_df=ledger_bal_v_amt[ledger_bal_v_amt['ExchType']==segment]
    segment_dfs.append(seg_df)
    seg_df_working=seg_df[['ClientCode','TotalVamt','Quarter_Dormancy']]
    seg_df_working_grouped=seg_df_working.groupby(['ClientCode','Quarter_Dormancy']).sum().reset_index()
    seg_df_working_grouped_pivot=seg_df_working_grouped.pivot(index='ClientCode',columns='Quarter_Dormancy',values='TotalVamt').reset_index().fillna(0)

    suffix='_'+segment.lower()+'_ledger_bal_total_v'
    seg_df_working_grouped_pivot=seg_df_working_grouped_pivot.add_suffix(suffix)
    seg_df_working_grouped_pivot.rename(columns={'ClientCode'+suffix:'ClientCode','Quarter_Dormancy'+suffix:'Quarter_Dormancy'},inplace=True)
    segment_pivot_dfs.append(seg_df_working_grouped_pivot)

    trimmed_col_list=['ClientCode']
    for col in seg_df_working_grouped_pivot.columns:
        col_yr=col[0:2]
        if col_yr in ["Y0","Y1","Y2","Y3"]:
            trimmed_col_list.append(col)
    seg_df_working_grouped_pivot_trimmed=seg_df_working_grouped_pivot[trimmed_col_list]
    segment_pivot_trimmed_dfs.append(seg_df_working_grouped_pivot_trimmed)


segment_df_dict=dict(zip(segments,segment_dfs))
segment_pivot_df_dict=dict(zip(segments,segment_pivot_dfs))
segment_pivot_df_trimmed_dict=dict(zip(segments,segment_pivot_trimmed_dfs))

ledger_bal_v_amt_merged = reduce(lambda  left,right: pd.merge(left,right,on=['ClientCode'],
                                            how='outer'), segment_pivot_trimmed_dfs)
ledger_bal_v_amt_merged_sorted = ledger_bal_v_amt_merged.reindex(sorted(ledger_bal_v_amt_merged.columns), axis=1)
cc_list=ledger_bal_v_amt_merged_sorted['ClientCode']
ledger_bal_v_amt_merged_sorted.pop('ClientCode')
ledger_bal_v_amt_merged_sorted.insert(0, 'ClientCode', cc_list)

ledger_bal_v_amt_merged_sorted['_ledger_bal_v_avg']= ledger_bal_v_amt_merged_sorted[total_cols].mean(axis=1)


ledger_bal_all_amt_merged_sorted=ledger_bal_bal_amt_merged_sorted.merge(ledger_bal_v_amt_merged_sorted,on='ClientCode',how='outer')
```
