## Import Libraries

```python
import pandas as pd
```

## Define files

```python
input_file_name="Mckinsey data 17th Aug for nomenclature.csv"
output_file_name="mckinsey_data_17_th_aug_with_nomenclature_debugged.csv"
df=pd.read_csv(input_file_name)
new_nomenclature=df
```

## New Nomenclature Rules (Post July 3rd)

```python
#app_new is for organic, and lssc app; Spot, Paid, Referral w. AppsFLyer ID
#web_new is for organic, and lssc web; Referral w.o AppsFLyer ID

#app_new_nomenclature start
registration_app_new_list=['Pre - Registration','Pre Registration','Mobile number enter',
                           'Mobile OTP enter','Email Address Enter']

pan_app_new_list=['Email OTP enter','Email Number Enter','Lead Registered','PAN Enter']

bank_details_app_new_list=['PAN Passed','Digilocker','Bank details']

personal_details_and_esign_app_new_list=['Bank details passed','Personal Details Passed','Selfie capture passed',
                                         'Signature Capture passed','Upload Your Documents',
                         'Upload Your Documents Passed','PendingEsign'
                                        ]

pending_for_activation_app_new_list=['Document Verification In Process (Final save)',
                                     'EsignDone','Esign Done','Move To Activation','In Activation Queue'
                                    ]
activated_app_new_list=['Client Activated']
#app_new_nomenclature end

#web_new_nomenclature start
registration_web_new_list=['Pre - Registration','Pre Registration','Mobile number enter',
                           'Mobile OTP enter','Email Address Enter','Lead Registered']

pan_web_new_list=['Email OTP enter','Email Number Enter','PAN Enter']

bank_details_web_new_list=['PAN Passed','Digilocker','Bank details']

personal_details_and_esign_web_new_list=['Bank details passed','Personal Details Passed','Selfie capture passed',
                                         'Signature Capture passed','Upload Your Documents',
                         'Upload Your Documents Passed','PendingEsign'
                                        ]
pending_for_activation_web_new_list=['Document Verification In Process (Final save)',
                                     'EsignDone','Esign Done','Move To Activation','In Activation Queue'
                                    ]

activated_web_new_list=['Client Activated']
#web_new_nomenclature_end

#converting all nomenclature conditions to lowercase
registration_app_new_list=list(map(str.lower, registration_app_new_list))
pan_app_new_list=list(map(str.lower, pan_app_new_list))
bank_details_app_new_list=list(map(str.lower, bank_details_app_new_list))
personal_details_and_esign_app_new_list=list(map(str.lower, personal_details_and_esign_app_new_list))
pending_for_activation_app_new_list=list(map(str.lower, pending_for_activation_app_new_list))
activated_app_new_list=list(map(str.lower, activated_app_new_list))

registration_web_new_list=list(map(str.lower, registration_web_new_list))
pan_web_new_list=list(map(str.lower, pan_web_new_list))
bank_details_web_new_list=list(map(str.lower, bank_details_web_new_list))
personal_details_and_esign_web_new_list=list(map(str.lower, personal_details_and_esign_web_new_list))
pending_for_activation_web_new_list=list(map(str.lower, pending_for_activation_web_new_list))
activated_web_new_list=list(map(str.lower, activated_web_new_list))


app_new_spot=new_nomenclature[new_nomenclature['Lead Source Category'].isin(['SPOT'])]
app_new_paid=new_nomenclature[new_nomenclature['Lead Source Category'].isin(['Paid','PAID'])]

new_organic=new_nomenclature[new_nomenclature['Lead Source Category'].isin(['Organic'])]

app_new_organic=new_organic[new_organic['Lead Source Sub Category'].isin(['MobileApp'])]
web_new_organic=new_organic[~new_organic['Lead Source Sub Category'].isin(['MobileApp'])]

new_referral=new_nomenclature[new_nomenclature['Lead Source Category'].isin(['Referral Lead'])]

#If appsflyer id exists --> app_referral, else web_referral
app_new_referral=new_referral[~new_referral['AppsFlyer ID'].isnull()]
web_new_referral=new_referral[new_referral['AppsFlyer ID'].isnull()]


#App_new work
nomenclature_col_for_app_new=[]

#spot
for row_pos in range(len(app_new_spot)):
    row=app_new_spot.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_app_new_list):
            nomenclature_col_for_app_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_app_new_list):
            nomenclature_col_for_app_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_app_new_list):
            nomenclature_col_for_app_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_app_new_list):
            nomenclature_col_for_app_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_app_new_list):
            nomenclature_col_for_app_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_app_new_list):
            nomenclature_col_for_app_new.append("activated")
    else:
        nomenclature_col_for_app_new.append("NONE")

app_new_spot['Nomenclature']=nomenclature_col_for_app_new

nomenclature_col_for_app_new=[]

#paid
for row_pos in range(len(app_new_paid)):
    row=app_new_paid.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_app_new_list):
            nomenclature_col_for_app_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_app_new_list):
            nomenclature_col_for_app_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_app_new_list):
            nomenclature_col_for_app_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_app_new_list):
            nomenclature_col_for_app_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_app_new_list):
            nomenclature_col_for_app_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_app_new_list):
            nomenclature_col_for_app_new.append("activated")
    else:
        nomenclature_col_for_app_new.append("NONE")


app_new_paid['Nomenclature']=nomenclature_col_for_app_new

nomenclature_col_for_app_new=[]

#organic
for row_pos in range(len(app_new_organic)):
    row=app_new_organic.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_app_new_list):
            nomenclature_col_for_app_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_app_new_list):
            nomenclature_col_for_app_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_app_new_list):
            nomenclature_col_for_app_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_app_new_list):
            nomenclature_col_for_app_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_app_new_list):
            nomenclature_col_for_app_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_app_new_list):
            nomenclature_col_for_app_new.append("activated")
    else:
        nomenclature_col_for_app_new.append("NONE")


app_new_organic['Nomenclature']=nomenclature_col_for_app_new


#Add Referral Part


nomenclature_col_for_app_new=[]
#referral
for row_pos in range(len(app_new_referral)):
    row=app_new_referral.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_app_new_list):
            nomenclature_col_for_app_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_app_new_list):
            nomenclature_col_for_app_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_app_new_list):
            nomenclature_col_for_app_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_app_new_list):
            nomenclature_col_for_app_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_app_new_list):
            nomenclature_col_for_app_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_app_new_list):
            nomenclature_col_for_app_new.append("activated")
    else:
        nomenclature_col_for_app_new.append("NONE")


app_new_referral['Nomenclature']=nomenclature_col_for_app_new

#web_new work

nomenclature_col_for_web_new=[]

#organic
for row_pos in range(len(web_new_organic)):
    row=web_new_organic.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_web_new_list):
            nomenclature_col_for_web_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_web_new_list):
            nomenclature_col_for_web_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_web_new_list):
            nomenclature_col_for_web_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_web_new_list):
            nomenclature_col_for_web_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_web_new_list):
            nomenclature_col_for_web_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_web_new_list):
            nomenclature_col_for_web_new.append("activated")
    else:
        nomenclature_col_for_web_new.append("NONE")


web_new_organic['Nomenclature']=nomenclature_col_for_web_new

#add referral part

nomenclature_col_for_web_new=[]

#referral
for row_pos in range(len(web_new_referral)):
    row=web_new_referral.iloc[row_pos,:]
    if row['Lead Sub Status'].lower() in (registration_web_new_list):
            nomenclature_col_for_web_new.append("registration")
    elif row['Lead Sub Status'].lower() in (pan_web_new_list):
            nomenclature_col_for_web_new.append("pan")
    elif row['Lead Sub Status'].lower() in (bank_details_web_new_list):
            nomenclature_col_for_web_new.append("bank_details")
    elif row['Lead Sub Status'].lower() in (personal_details_and_esign_web_new_list):
            nomenclature_col_for_web_new.append("personal_details_and_docs_esign")
    elif row['Lead Sub Status'].lower() in (pending_for_activation_web_new_list):
            nomenclature_col_for_web_new.append("pending_for_activation")
    elif row['Lead Sub Status'].lower() in (activated_web_new_list):
            nomenclature_col_for_web_new.append("activated")
    else:
        nomenclature_col_for_web_new.append("NONE")


web_new_referral['Nomenclature']=nomenclature_col_for_web_new


#SPOT_Merged
spot_final=pd.concat([app_new_spot])
spot_final['LSC_Categorization']=["spot"]*len(spot_final)

#PAID_Merged
paid_final=pd.concat([app_new_paid])
paid_final['LSC_Categorization']=["paid"]*len(paid_final)

#Organic_Merged
organic_final=pd.concat([app_new_organic,web_new_organic])
organic_final['LSC_Categorization']=["organic"]*len(organic_final)

#referral_Merged
referral_final=pd.concat([app_new_referral,web_new_referral])
referral_final['LSC_Categorization']=["referral"]*len(referral_final)


all_final=pd.concat([spot_final,paid_final,organic_final,referral_final])
all_final
```

## Saving File

```python
all_final=all_final[all_final['Nomenclature'].isin(['registration','pan','bank_details','personal_details_and_docs_esign'])]
all_final.to_csv(output_file_name,index=False)
```
